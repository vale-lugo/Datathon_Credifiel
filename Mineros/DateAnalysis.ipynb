```python
dfDaysOfWeek = listCobro.merge(listCobroEmisora, on="idListaCobro", how="left")
dfDaysOfWeek["fechaCreacionLista"] = pd.to_datetime(dfDaysOfWeek['fechaCreacionLista'], format='%d/%m/%Y %I:%M%p')
dfDaysOfWeek["fechaEnvioCobro"] = pd.to_datetime(dfDaysOfWeek['fechaEnvioCobro'], format="%d/%m/%Y" )
dfDaysOfWeek["createdAt"] = dfDaysOfWeek["fechaCreacionLista"].dt.day_name()
dfDaysOfWeek["sendIt"] = dfDaysOfWeek["fechaEnvioCobro"].dt.day_name()
dfDaysOfWeek
```

```python
dayPayed = listCobro2024.merge(dfDaysOfWeek, on="idListaCobro", how="left")
#dayPayed["count"] = 1
dayPayed
```

```python
dfEmi = dayPayed.groupby(['sendIt', 'idRespuestaBanco', "idEmisora"], as_index=False).agg({
    "montoExigible": "sum",
    "montoCobrar": 'sum',
    'montoCobrado': 'sum', 
    "count":"count",
    "idBanco_x":"first",
})
dfEmi["Monto"] = dfEmi["montoCobrado"] - dfEmi["montoCobrar"]
day_totals = dfEmi.groupby('sendIt')['count'].transform('sum')

# Calculate percentage per row
dfEmi['count_pct'] = dfEmi['count'] / day_totals * 100
dfEmi['count_pct'] = dfEmi['count_pct'].round(2)
dfEmi["Monto"] = dfEmi["Monto"].round(2)
#dfEmi = dfEmi[dfEmi["idRespuestaBanco"] == 0.0]
dfEmi = dfEmi[dfEmi["count_pct"] > 1]

def highlight_cells(val):
    return 'color: red;' if val > .259 else ''


styled = dfEmi.style\
    .map(highlight_cells, subset=['count_pct'])
styled
```

```python
numOrdersPerBankE = dayPayed.groupby(["idBanco_y", "idRespuestaBanco"]).agg({"count": "count"}).reset_index()
numOrdersPerBankE = numOrdersPerBankE[numOrdersPerBankE["idRespuestaBanco"] == 0.0]
numOrdersPerBankE  = numOrdersPerBankE.drop(columns={"idRespuestaBanco"})
numOrdersPerBankE = numOrdersPerBankE.rename(columns={"count":"effective"})
numOrdersPerBankE
```

```python
numOrdersPerBank = dayPayed.groupby(["idBanco_y"]).agg({"count":"count"})
numOrdersPerBank = numOrdersPerBank.merge(numOrdersPerBankE, on="idBanco_y", how="left")
numOrdersPerBank["count"] = (numOrdersPerBank["count"] / 5900894)*100
numOrdersPerBank["effective"] = (numOrdersPerBank["effective"] / 5900894)*100
numOrdersPerBank = numOrdersPerBank.reset_index()
```

```python
# Plot
plt.figure(figsize=(8, 5))

bars = plt.bar(numOrdersPerBank["idBanco_y"].astype(str), numOrdersPerBank["effective"], color='lightgreen')


plt.xlabel("idBanco")
plt.ylabel("% Transacciones Efectivas")
plt.title("% de Transacciones por Banco Efectivas")
plt.tight_layout()
plt.savefig("Transaction Count per Day Envio.jpg", format="jpg")
plt.show()
```

```python
# Plot
plt.figure(figsize=(8, 5))

bars = plt.bar(numOrdersPerBank["idBanco_y"].astype(str), numOrdersPerBank["count"], color='skyblue')

# Agrega etiquetas encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)),
             ha='center', va='bottom', fontsize=9)

plt.xlabel("idBanco")
plt.ylabel("% Transacciones")
plt.title("% de Transacciones por Banco")
plt.tight_layout()
plt.savefig("Transaction Count per Day Envio.jpg", format="jpg")
plt.show()
```
